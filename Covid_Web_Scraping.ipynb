```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup
```

```python
soup = soup.text
```

```python
!pip install lxml
```

```python
code = BeautifulSoup(soup, "lxml")
```

```python
code = code.table
```

```python
data = []
tags = code.find_all('tr')
for tag in tags:
  x = tag.text
  data.append(x)

data
```

```python
datas = []
for i in data:
  datas.append(i.split('\n')[1:-1])
```

```python
cleaned_data = [i for i in datas if i[0] != '']
```

```python
import csv
file = open('covid_data.csv','w')
x = csv.writer(file)
for i in cleaned_data:
  x.writerow(i)

file.close()
```

```python
import pandas as pd
df = pd.read_csv('/content/covid_data.csv')
df
```

```python
country = list(df['Country,Other'])[0:10]
total_cases = list(df['TotalCases'])[0:10]
total_cases
cleaned_cases = []
for cases in total_cases:
  x = int(cases.replace(',',''))
  cleaned_cases.append(x)

cleaned_cases
```

```python
import plotly.graph_objects as go
#x = country
# y = cleaned_cases

fig = go.Figure([go.Bar(x=country, y = cleaned_cases)])
fig.show()
```

```python
import plotly.graph_objects as go

labels = country
values = cleaned_cases

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```
